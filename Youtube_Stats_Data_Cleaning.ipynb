```python
import numpy as np
import pandas as pd
from scipy import stats
from IPython.display import display, Markdown
```

```python
data = pd.read_csv("INvideos.csv")
data
```

```python
#Display basic information about the dataset
display(Markdown("**INFO About Data**"))
print(data.info())
display(Markdown("**Describing About Data**"))
print(data.describe())
display(Markdown("**Dataset Datatype**"))
print(data.dtypes)
```

```python
# Coverting all columns name first letter in capital
data.columns = data.columns.str.title()
data
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Drop rows where the 'description' column has null (NaN) values
data = data.dropna(subset=['description'])
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Standardizing some column
data['Publish_Time'] = pd.to_datetime(data['Publish_Time'])
data['Title'] = data['Title'].str.title()
data['Description'] = data['Description'].str.title()
data['Views'] = data['Views'].astype(int)
data['Likes'] = data['Likes'].astype(int)
data['Dislikes'] = data['Dislikes'].astype(int)
```

```python
data
```

```python

def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)  # 25th percentile
        Q3 = df[column].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range
        
        # Define the lower and upper bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Remove rows where the values are outside of the bounds
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df

# Specify the numeric columns where outliers need to be removed
numeric_columns = [ 'views','likes', 'dislikes', 'comment_count']

# Remove outliers from the dataset
data_cleaned = remove_outliers_iqr(data, numeric_columns)
```
